```python
weight = int(input('몸무게 : '))
height = int(input('키 : ')) / 100

bmi = weight / (height*height)
print('BMI =', bmi)

# BMI에 따른 비만도 판단 조건식

if bmi < 18.5:
    print('저체중')
elif bmi <= 23:
    print('정상')
elif bmi <= 25:
    print('과체중')
elif bmi <= 30:
    print('비만')
else:
    print('고도비만')
```

```python
import random
scores = []
for i in range(200):
    scores.append(random.randint(1, 100))

cnt = 0
for score in scores:
    if score >= 90:
        cnt += 1
        print(score)
print('총개수',cnt)
```
