```python
import matplotlib.pyplot as plt

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 한글폰트 적용
plt.rc('font',family = 'Malgun Gothic')

# 제목
plt.title('월별 판매 실적')
plt.plot(month, sales,color='b',label='단위:개수')
plt.legend(loc='lower right') # 위치를 정해준다 (오른쪽 아래)
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')

sales = [1,5,7,3,7]
month = [3,4,5,6,7]

plt.barh(month,sales)
plt.show()
```
