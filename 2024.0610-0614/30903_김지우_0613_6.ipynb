```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

sales = [1, 5, 7, 3, 7]
month = [3, 4, 5, 6, 7]

plt.title('월별 판매 실적')
plt.bar(month, sales, color='b')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

sales = [1, 5, 7, 3, 7]
month = [3, 4, 5, 6, 7]

plt.title('월별 판매 실적')
plt.barh(month, sales, color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

sales = [1, 5, 7, 3, 7]

plt.title('월별 판매 실적')
plt.bar(range(3, 8), sales, color = 'orange')
plt.show()
```
