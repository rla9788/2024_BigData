```python
import matplotlib.pyplot as plt

sales = [1,5,7,3,7]
month = ['mar','par','may','jun','jul']

# 그래프 해상도 설정
plt.figure(dpi = 150)

# 한글 폰트
plt.rc('font',family='Malgun Gothic')

# 그래프 제목 넣기 plt.title
plt.title('월별 판매 실적')

plt.plot(month, sales, color = 'red') # 그래프 색 바꾸기 color = "색이름"
plt.show()
```
