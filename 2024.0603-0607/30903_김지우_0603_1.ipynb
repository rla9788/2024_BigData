```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 그래프의 해상도 설정 dpi .figure
plt.figure(dpi = 100)

# 제목 title
plt.title('월별 판매 실적')

# 꺾은선형 그래프 그리기 .plot
# 그래프 범례 .label
plt.plot(month, sales, color = 'y', label = '단위:개수')

# 범례 보이게 설정 .lower right
plt.legend(loc = 'lower right')

# x축 y축 xlabel, ylabel
plt.xlabel('월')
plt.ylabel('매출')

# 그래프 보이기
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

plt.figure(dpi = 100)

# data
sales = [1,5,7,3,7]
month = [3,4,5,6,7]

# 막대 그래프 plt.bar()
plt.title('월별 판매 실적')
plt.bar(month, sales, color = 'b')

plt.show()
```
