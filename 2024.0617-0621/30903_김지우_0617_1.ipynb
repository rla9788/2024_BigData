```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')

sales = [1,5,7,3,7]
month = [3,4,5,6,7]

plt.title('월별 판매 실적')

plt.barh(range(3, 8), sales, color = 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')

# data 혈액형 수
b_type = [25, 19, 37, 11]
b_name = ['A형','B형','O형','AB형']

plt.title('혈액형 분포')

# 원형차트 pie 범례 labels = [] autopct = ''
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')

# data 국어
kor = [80,20,50,20,10,50,60,30,40]
eng = [90,40,10,40,60,20,50,70,90]

# 산점도 scatter option color, 투명도 alpha
plt.scatter(kor, eng, color = 'r', alpha = 0.7)

# 축제목 x, y
plt.xlabel('국어점수')
plt.ylabel('영어점수')

plt.show()
```
