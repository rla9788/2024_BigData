```python
import matplotlib.pyplot as plt
import random

plt.rc('font',family='Malgun Gothic')

height = []
weight = []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
# 산점도
plt.scatter(height, weight, c = height, cmap = 'RdPu')
plt.colorbar(label = '키')

plt.title('키와 몸무게의 상관관계')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

plt.rc('font',family='Malgun Gothic')

height, weight, score = [],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
    
# 산점도
plt.scatter(height, weight, s = score, c = score, cmap = 'rainbow', alpha = 0.7)
plt.colorbar(label = '체육 실기 점수')
plt.title('키와 몸무게의 상관관계')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

## 그래프 겹쳐 그리기
### 꺾은선 그래프 겹쳐 그리기

```python
import matplotlib.pyplot as plt

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.plot(singer,week1, c = 'hotpink', label = '첫째주')
plt.plot(singer,week2, c = 'royalblue', label = '둘째주')
plt.title('오디션 프로그램 득표 현황')
plt.legend()
plt.show()
```
